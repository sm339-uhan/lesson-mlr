## Example

How is an NFL team's winning percentage related to its offensive and defensive performance?
The dataset `NFLStandings2016` from `Stat2Data` contains the records for all NFL teams during the 2016 regular season.
`WinPct` is the winning percentage, `PointsFor` is the total number of points scored, and `PointsAgainst` is the total number of points allowed.

First, let's load the dataset.

```r
library(Stat2Data)
data(NFLStandings2016)
```

Next, let's fit the multiple linear regression model, with $\mathit{WinPct}$ as the response variable, and $\mathit{PointsFor}$ and $\mathit{PointsAgainst}$ as the explanatory variables.

```r
fit <- lm(WinPct ~ PointsFor + PointsAgainst, data = NFLStandings2016)
summary(fit)
```

The following code creates the residuals vs. fitted values plot and the Normal Q-Q plot of the residuals:

```r
plot(fit, which=1)
plot(fit, which=2)
```
